```python
#Taking f1--f166 as independent...
X= data.iloc[:,2:-1].values
y=data.iloc[:,-1].values
```

```python
#Spliting dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
# Scaling Of Independent Variables
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
models=[
{
    "label": "KNN",
    "model": KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2),
},
{
    "label": "Decision Tree",
    "model": DecisionTreeClassifier(criterion = 'entropy', random_state = 0),
},
{
    "label": "Random Forest",
    "model": RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0),
},
{
    "label": "Naive Bayes",
    "model": GaussianNB()
},
{
    "label": "SVC RBF",
    "model": SVC(kernel = 'rbf', random_state = 0)      
},
{
    "label": "SVC LINEAR",
    "model": SVC(kernel = 'linear', random_state = 0)
},
{
    "label": "SVC POLYNOMIAL",
    "model": SVC(kernel = 'poly', random_state = 0)
},
{
    "label": "LOGISTIC",
    "model": LogisticRegression(random_state = 0)
},

]

for m in models:
    model= m["model"]
    model.fit(X_train, y_train)
    y_pred= model.predict(X_test)
    cm= confusion_matrix(y_test, y_pred)
    print("Confusion Matrix" +m["label"]+":" ,cm)
    print("Accuracy Score " +m["label"]+":" , accuracy_score(y_test,y_pred))
    print("Precision Score " +m["label"]+":", precision_score(y_test,y_pred, average="weighted"))
    print("Recall Score " +m["label"]+":", recall_score(y_test,y_pred, average="weighted"))
    print()
```

```python
def mole_name1(row):
    return row["molecule_name"].rsplit('-',1)[0]
def mole_name2(row):
    return row["molecule_name"].rsplit("-",1)[-1]                                    
data["Musk_NonMusk"]= data.apply(mole_name1, axis=1)
data["catagory"]= data.apply(mole_name2, axis=1)                                
```
